```python
import cv2
import matplotlib.pyplot as plt
```

```python
path_2=r"/content/img2.jpg"
path_3=r"/content/img3.jpg"
```

```python
img_2=cv2.imread(path_2,0)
```

```python
img_3=cv2.imread(path_3,0)
```

```python
alpha=0.5
beta=1-alpha
```

```python
def gemma_adjust(gemma):
  y=cv2.addWeighted(img_2,alpha,img_3,beta,gemma)
  #cv2.imwrite('gemma_adjust.jpg',y)
  return y
```

```python
g_nega_50=gemma_adjust(-50)
g_nega_10=gemma_adjust(-10)
g_10=gemma_adjust(10)
g_50=gemma_adjust(50)
```

```python
def weight_adjust(alpha_2):
  beta_2=1-alpha_2
  gemma_2=1
  y=cv2.addWeighted(img_2,alpha_2,img_3,beta_2,gemma_2)
  #cv2.imwrite('gemma_adjust.jpg',y)
  return y
```

```python
a_2=weight_adjust(0.2)
a_4=weight_adjust(0.4)
a_6=weight_adjust(0.6)
a_8=weight_adjust(0.8)
```

```python
plt.subplot(241)
plt.imshow(g_nega_50)
plt.axis('off')

plt.subplot(242)
plt.imshow(g_nega_10)
plt.axis('off')

plt.subplot(243)
plt.imshow(g_10)
plt.axis('off')

plt.subplot(244)
plt.imshow(g_50)
plt.axis('off')

plt.subplot(245)
plt.imshow(a_2)
plt.axis('off')

plt.subplot(246)
plt.imshow(a_4)
plt.axis('off')

plt.subplot(247)
plt.imshow(a_6)
plt.axis('off')

plt.subplot(248)
plt.imshow(a_8)
plt.axis('off')
```
